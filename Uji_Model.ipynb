**Load Model & Tokenizer**

```python
from tensorflow.keras.models import load_model
import pickle

MODEL_PATH = "/content/drive/MyDrive/Kecerdasan Buatan Lanjut/UAS/bi_gru_model.h5"
TOKENIZER_PATH = "/content/drive/MyDrive/Kecerdasan Buatan Lanjut/UAS/tokenizer.pkl"

model = load_model(MODEL_PATH)

with open(TOKENIZER_PATH, "rb") as f:
    tokenizer = pickle.load(f)

print("Model & tokenizer siap digunakan")
```

**Preprocessing**

```python
import re
from tensorflow.keras.preprocessing.sequence import pad_sequences

MAX_LEN = 100

def clean_text(text):
    text = text.lower()
    text = re.sub(r'http\S+|www\S+', '', text)
    text = re.sub(r'[^a-z0-9\s]', ' ', text)
    text = re.sub(r'(.)\1{2,}', r'\1', text)
    text = re.sub(r'\s+', ' ', text).strip()
    return text
```

**Fungsi Prediksi**

```python
def predict_text(text):
    text = clean_text(text)
    seq = tokenizer.texts_to_sequences([text])
    pad = pad_sequences(seq, maxlen=MAX_LEN, padding='post')

    prob = model.predict(pad, verbose=0)[0][0]
    label = "Judol" if prob >= 0.5 else "Non Judol"

    return label, prob
```

**USER**

```python
while True:
    text = input("\nMasukkan teks (ketik 'exit' untuk keluar): ")
    if text.lower() == "exit":
        break

    label, prob = predict_text(text)
    print("Prediksi     :", label)
    print("Probabilitas :", f"{prob:.4f}")
```
